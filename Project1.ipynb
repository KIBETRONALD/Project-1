```python
import matplotlib.pyplot as plt

Ova_saleprice = df1[['OverallCond', 'SalePrice']]

Total_salesprice = {}
for i in Ova_saleprice['OverallCond'].unique():
    condition_values = Ova_saleprice.loc[Ova_saleprice['OverallCond'] == i, 'SalePrice'].tolist()
    Total_salesprice[i] = sum(condition_values)

sale_price = Total_salesprice.items()
Sorted_salesprice = sorted(sale_price, key=lambda x: x[0])
x = [item[0] for item in Sorted_salesprice]
y = [item[1] for item in Sorted_salesprice]

# plotting a bar plot for quantity against sales
fig, ax = plt.subplots()
bars = ax.bar(x, y)
ax.set_ylabel('Total salesprice')
ax.set_xlabel('Overallcond')
ax.set_title('Salesprice distribution per Overallcond')

# Display values on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

plt.show()
```

```python
import matplotlib.pyplot as plt
#Separate the data into subsets based on OverallCond, then demonstrate how this split impacts the distribution of SalePrice.
# Assuming df1 is your DataFrame
Ova_saleprice = df1[['OverallCond', 'SalePrice']]

Total_salesprice = {}
for i in Ova_saleprice['OverallCond'].unique():
    condition_values = Ova_saleprice.loc[Ova_saleprice['OverallCond'] == i, 'SalePrice'].tolist()
    Total_salesprice[i] = sum(condition_values)

sale_price = Total_salesprice.items()
Sorted_salesprice = sorted(sale_price, key=lambda x: x[0])
x = [item[0] for item in Sorted_salesprice]
y = [item[1] for item in Sorted_salesprice]

# plotting a horizontal bar plot for quantity against sales
fig, ax = plt.subplots()
bars = ax.barh(x, y)  # Use barh instead of bar
ax.set_xlabel('Total salesprice')
ax.set_ylabel('Overallcond')
ax.set_title('Salesprice distribution per Overallcond')

# Display values on the right of each bar
for bar in bars:
    xval = bar.get_width()
    plt.text(xval, bar.get_y() + bar.get_height()/2, round(xval, 2), ha='left', va='center')

# Adjust y-axis and x-axis limits for horizontal bar plot
#ax.set_ylim(min(x) - 0.5, max(x) + 0.1)  # You can adjust the margins as needed
ax.set_xlim(0, max(y) + 10000)  # You can adjust the upper limit of x-axis as needed

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
# Extract the relevant columns
selected_columns = ['SalePrice', 'TotRmsAbvGrd', 'OverallCond']
selected_data = df[selected_columns]

# Summary Statistics
summary_statistics = selected_data.describe().T

# Visualizations
# Histograms for SalePrice, TotRmsAbvGrd
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

selected_data['SalePrice'].plot(kind='hist', bins=20, ax=axes[0, 0], edgecolor='black')
axes[0, 0].set_title('Distribution of SalePrice')
axes[0, 0].set_xlabel('SalePrice')

selected_data['TotRmsAbvGrd'].plot(kind='hist', bins=10, ax=axes[0, 1], edgecolor='black')
axes[0, 1].set_title('Distribution of TotRmsAbvGrd')
axes[0, 1].set_xlabel('TotRmsAbvGrd')

# Boxplot for OverallCond
selected_data.boxplot(column='OverallCond', ax=axes[1, 0])
axes[1, 0].set_title('Boxplot of OverallCond')

# Scatter plot of TotRmsAbvGrd vs SalePrice
axes[1, 1].scatter(selected_data['TotRmsAbvGrd'], selected_data['SalePrice'], alpha=0.5)
axes[1, 1].set_title('Scatter plot of TotRmsAbvGrd vs SalePrice')
axes[1, 1].set_xlabel('TotRmsAbvGrd')
axes[1, 1].set_ylabel('SalePrice')

plt.tight_layout()
plt.show()

# Interpretive Text
print("Summary Statistics:")
print(summary_statistics)

print("\nInterpretation:")
print("1. SalePrice Distribution: The histogram shows the distribution of SalePrice. It appears to be right-skewed.")
print("2. TotRmsAbvGrd Distribution: The histogram shows the distribution of TotRmsAbvGrd.")
print("3. OverallCond Boxplot: The boxplot displays the distribution of OverallCond. Outliers may be present below the lower whisker.")
print("4. Scatter plot: The scatter plot depicts the relationship between TotRmsAbvGrd and SalePrice.")
```
